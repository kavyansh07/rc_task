```python
import cv2
import numpy as np
```

```python
from google.colab.patches import cv2_imshow
```

```python
aiml=cv2.imread("AIML.jpg")
cv2_imshow(aiml)
```

# Adding rectangle and polygon on the white space

```python
aiml=cv2.resize(aiml,(800,500)) #Resizing the image in order to take proper cordinates
cv2_imshow(aiml)
```

```python
aiml_new=cv2.rectangle(aiml,pt1=(600,100),pt2=(700,200), color=(0,250,0),thickness=5,lineType=16) # Adding square in the white space
```

```python
cv2_imshow(aiml_new)
```

```python
# Adding 3 sided polygon or triangle in the white space
aiml_1=cv2.polylines(aiml_new,pts=[np.array([[650,300],[600,400],[700,400]])],isClosed=True, color=(0,255,0), thickness=5, lineType=16)
cv2_imshow(aiml_1)
```

# Gray, rgb, hsv



```python
aiml_1Gray = cv2.cvtColor(aiml_1, cv2.COLOR_BGR2GRAY)
cv2_imshow(aiml_1Gray)
```

```python
aiml_1rgb = cv2.cvtColor(aiml_1, cv2.COLOR_BGR2RGB)
cv2_imshow(aiml_1rgb)
```

```python
aiml_1hsv = cv2.cvtColor(aiml_1, cv2.COLOR_BGR2HSV)
cv2_imshow(aiml_1hsv)
```

# Gaussian Blur

```python
aiml_blur = cv2.GaussianBlur(aiml_1, (0, 0), 5)
cv2_imshow(aiml_blur)
```
